# Student Lifestyle Analysis

## Objective

This project aims to analyze how lifestyle factors such as alcohol consumption, social activities, and attendance affect student academic performance.

## 1. Data Exploration

```python
import pandas as pd

df = pd.read_csv("student_data.csv")
df.head()
```

## 2. Exploratory Data Analysis (EDA)

```python
import matplotlib.pyplot as plt

df['G3'].hist()
plt.title("Final Grade Distribution")
plt.show()
```

## Analysis 1：Absences vs Grades

```python
import matplotlib.pyplot as plt
plt.scatter(df['absences'], df['G3'])
plt.xlabel("Absences")
plt.ylabel("Final Grade (G3)")
plt.title("Absences vs Final Grade")
plt.show()
```

## Insight 1: Absences and Performance

There is a slight negative relationship between absences and final grades.

Students with higher absence levels tend to have lower academic performance, although the relationship is not strongly linear.

## Analysis 2 : Alcohol vs Grades

```python
df.groupby('Dalc')['G3'].mean().plot(kind='bar')
plt.title("Workday Alcohol Consumption vs Grade")
plt.xlabel("Alcohol Level")
plt.ylabel("Average Grade")
plt.show()
```

## Insight 2: Alcohol Consumption

Higher alcohol consumption during weekdays appears to be associated with slightly lower academic performance.

This suggests that unhealthy habits may negatively impact student outcomes.

## Analysis 3 : Going out vs Grades

```python
df.groupby('goout')['G3'].mean().plot(kind='bar')
plt.title("Going Out vs Grade")
plt.xlabel("Going Out Frequency")
plt.ylabel("Average Grade")
plt.show()
```

## Insight 3: Social Life and Performance

Students who go out more frequently tend to have slightly lower grades.

However, moderate social activity does not appear to significantly harm performance.

## Analysis 4 : Study Times vs Grades

```python
plt.scatter(df['studytime'], df['G3'])
plt.xlabel("Study Time")
plt.ylabel("Final Grade")
plt.title("Study Time vs Final Grade")
plt.show()
```

## Insight 4: Study Time

There is a weak positive relationship between study time and academic performance.

This indicates that studying contributes to better outcomes, but it is not the only determining factor.

## Analysis 5 : Classification between Pass and Fail

```python
df['pass'] = df['G3'] >= 10

df.groupby('pass')['G3'].count()
```

## Insight 5: Pass vs Fail

Most students pass the course, but those with poor lifestyle habits are more likely to fall into the failing group.

## Conclusion

This analysis shows that lifestyle factors such as attendance, alcohol consumption, and social activity have noticeable effects on student performance.

While study time plays a role, behavioral factors appear to be equally important.

These findings suggest that improving daily habits may help students achieve better academic outcomes.
